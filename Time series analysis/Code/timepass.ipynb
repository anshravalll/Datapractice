```python
df = pd.read_csv("DailyDelhiClimateTrain.csv")
```

```python
#Recreate the head of this particular dataframe with 2 ways. 1 - through making a dicitonary dataframe and 2 - using looping.
for row in df.itertuples():
    print(row.date)
    
```

```python
def do_something(idk):
    return pd.to_datetime(idk["date"]).strftime("%d-%m-%Y")
    
df["date"] = df.apply(do_something, axis = 1)
df["date"]
```

```python
# Compare to rolling standard deviation
df["change_categorization"] = df["humidity"].rolling(window=7).mean().fillna(0).diff().fillna(0).apply(
    lambda x: "More" if x > 0 else "Less"  # Or set a specific threshold
)
```

```python
df["change_categorization"]
```

```python
#Change_categorization using conditions
conditions = [
    (df["meantemp"] > 35, "Extreme Heat"),
    (df["wind_speed"] > 15, "Windy"),
    (df["humidity"] > 90, "Very Humid"),
    (df["meanpressure"] < 1000, "Low Pressure")
]

for condition, label in conditions:
    df.loc[condition, "weather_status"] = label
```
